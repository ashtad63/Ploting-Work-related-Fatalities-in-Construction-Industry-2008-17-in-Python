```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white", {'axes.grid': True})
#sns.set_palette("Paired")
%matplotlib inline
```

```python
df = pd.read_excel (r'C:\Users\Ashtad\Desktop\Fatal Injuries (count).xlsx')
```

```python
df.index = df['Year']
df.drop('Year', axis=1, inplace=True)
df.head()
```

```python
plt.figure(figsize=(10,5), dpi=300)
sns.lineplot(data=df, x=df.index, y='Natural resources and mining', linewidth=2.5, marker='o', color='#006400');
sns.lineplot(data=df, x=df.index, y='Construction', linewidth=2.5, marker='o', color='#e31a1c');
sns.lineplot(data=df, x=df.index, y='Manufacturing', linewidth=2.5, marker='o', color= '#33a02c');
sns.lineplot(data=df, x=df.index, y='Trade, transportation and utilities', linewidth=2.5, marker='o', color='#b03060');
sns.lineplot(data=df, x=df.index, y='Information', linewidth=2.5, marker='o', color='#1f78b4');
sns.lineplot(data=df, x=df.index, y='Financial activities', linewidth=2.5, marker='o', color='#fb9a99');
sns.lineplot(data=df, x=df.index, y='Professional and business services', linewidth=2.5, marker='o');
sns.lineplot(data=df, x=df.index, y='Educational and health services', linewidth=2.5, marker='o', color='#9A6324');
sns.lineplot(data=df, x=df.index, y='Leisure and hospitality', linewidth=2.5, marker='o', color='#ff7f00');
sns.lineplot(data=df, x=df.index, y='Other services', linewidth=2.5, marker='o', color='#00bfff');

plt.xlabel("");
plt.ylabel("Recorded Fatal Injuries");
plt.legend(labels=['Natural resources and mining', 'Construction', 'Manufacturing',
                  'Trade, transportation and utilities', 'Information', 'Financial activities',
                  'Professional and business services', 'Educational and health services',
                  'Leisure and hospitality', 'Other services'], 
           loc=0, frameon=False, bbox_to_anchor=(1,1));
plt.title('Number of work-related fatalities and catastrophes recorded in the U.S. (2010-2017)')
sns.despine()
```
