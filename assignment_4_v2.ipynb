```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style("white", {'axes.grid': True})
#sns.set_palette("Paired")

%matplotlib inline
```

```python
df = pd.read_excel (r'C:\Users\Ashtad\Desktop\Fatal Injuries (rate).xlsx')
```

```python
df.index = df['Year']
df.drop('Year', axis=1, inplace=True)
df.head()
```

```python
plt.figure(figsize=(9,5), dpi=300)
sns.lineplot(data=df, x=df.index, y='Agriculture, forestry, fishing, and hunting', linewidth=2.5, marker='o', color='#006400');
sns.lineplot(data=df, x=df.index, y='Mining, quarrying, and oil and gas extraction', linewidth=2.5, marker='o', color='#ff7f00');
sns.lineplot(data=df, x=df.index, y='Construction', linewidth=2.5, marker='o', color='#e31a1c');
sns.lineplot(data=df, x=df.index, y='Manufacturing', linewidth=2.5, marker='o', color='#33a02c');

plt.xticks(np.arange(2008,2018,1))
plt.xlabel("");
plt.ylabel("Fatality rate per 100,000 full-time equivalent workers");
sns.despine()

plt.title('Work-related fatalities in construction and other goods producing industries in the U.S. (2008-17)', 
          fontdict={'weight':'bold'}, loc='left', pad=10)

plt.legend(labels=['Agriculture, forestry, fishing, and hunting', 'Mining, quarrying, and oil and gas extraction', 
                   'Construction', 'Manufacturing'], 
           loc=0, frameon=False, bbox_to_anchor=(1,1));
```
